```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

# Pre-Processing Dataset

```python
mmd = pd.read_csv("mmd.csv")
mmd.head()
```

```python
a = []
b = []
c= []
t = []
for i in mmd["Vehicle Make"].unique():
    #pass_model['Vehicle Make'][j] = i 
    a.append((sum(mmd["PASS"][mmd["Vehicle Make"].eq(i)])/
              sum(mmd["Total"][mmd["Vehicle Make"].eq(i)]))*100) #storing pass avg
    b.append(i)  # storing company name
    c.append(mmd["Vehicle Model"][mmd["Vehicle Make"].eq(i)].count())
    t.append(sum(mmd["Total"][mmd["Vehicle Make"].eq(i)]))
    #g.append(sum(mmd["OTHER"][mmd["Vehicle Make"].eq(i)]))
```

```python
pass_model = pd.DataFrame(columns=['Vehicle Make', 'Pass_Average(%)', 'Model_Count'])
```

```python
pass_model["Pass_Average(%)"] = a
pass_model["Vehicle Make"] = b
pass_model['Model_Count'] = c
pass_model["Total cars tested"] = t
pass_model
```

# Graph for Vehicle_make vs Different Attributes

```python
# Create figure
fig = go.Figure()


df = pass_model.sort_values(by=["Pass_Average(%)"], ascending = False)[:20].sort_values(by=["Pass_Average(%)"])
# Add traces
fig.add_trace(
    go.Bar(x=df["Pass_Average(%)"], 
        y=df["Vehicle Make"],orientation='h', name = "Pass_Average(%)", visible = False)
)
                                                                                            
df1 = pass_model.sort_values(by=["Model_Count"], ascending = False)[:20].sort_values(by=["Model_Count"])
                                                                                            
fig.add_trace(
    go.Bar(x=df1["Model_Count"], 
        y=df1["Vehicle Make"],orientation='h', name = "Model_Count",visible = False)
)

df2 = pass_model.sort_values(by=["Total cars tested"], ascending = False)[:20].sort_values(by=["Total cars tested"])

fig.add_trace(
    go.Bar(x=df2["Total cars tested"], 
        y=df2["Vehicle Make"],orientation='h', name = "Total cars tested",visible = False)
)

Pass_Average = [dict(x= df["Pass_Average(%)"],
                 y = df["Vehicle Make"],
                            xref="x", yref="y",
                 ax=0, ay=-40
                            )]
Model_Count = [dict(x= df1["Model_Count"],
                 y = df1["Vehicle Make"],
                            xref="x", yref="y",
                 ax=0, ay=-40
        
                            )]

Total_cars_tested = [dict(x= df2["Total cars tested"],
                 y = df2["Vehicle Make"],
                            xref="x", yref="y",
                 ax=0, ay=-40
        
                            )]
fig.update_layout(
    updatemenus=[
        dict(buttons=list([
            dict(label="None",
                     method="update",
                     args=[{"visible": [False, False,False]},
                           {"title": "Select x axis from the Drop Down menu",
                            "annotations": []}]),
            dict(label="Pass_Average",
                 method="update",
                 args=[{"visible": [True, False,False]},{
                            "title": "Vehicle Make vs Pass_Average(%) (Top 20)","annotations": Pass_Average}]),
            dict(label="Model Count",
                 method="update",
                 args=[{"visible": [False, True,False]},{
                            "title": "Vehicle Make vs Model_Count (Top 20)","annotations": Model_Count}]),
            dict(label="Total Cars Tested",
                 method="update",
                 args=[{"visible": [False, False,True]},{
                            "title": "Vehicle Make vs Total Cars Tested (Top 20)","annotations": Total_cars_tested}]),
        ]))]

)

fig.update_layout(
    #xaxis_title="Vehicle Make",
    yaxis_title="Vehicle Make",
    title_text="Select the X-axis from the drop down menu",
    showlegend=False,
)
fig.show()
fig.write_html("vehicle_make_vs_attrs.html")
```

```python
import plotly
print(plotly.offline.plot(fig, include_plotlyjs=False, output_type='div'))
```

# Analysis of Number of models tested with respect to their year of Built

```python
year_pass = pd.DataFrame(columns=['Year of Birth', 'Pass Total'])
d=[]
e=[]
f=[]
for i in mmd["Year Of Birth"].unique():
    d.append((sum(mmd["PASS"][mmd["Year Of Birth"].eq(i)])/
              sum(mmd["Total"][mmd["Year Of Birth"].eq(i)]))*100)
    e.append(i)
    f.append(sum(mmd["Total"][mmd["Year Of Birth"].eq(i)]))
year_pass = pd.DataFrame(columns=['Year Of Birth', 'Pass_Average(%)'])
year_pass["Year Of Birth"] = e
year_pass["Pass_Average(%)"] = d
year_pass["Total Models Tested"] = f
year_pass['Year Of Birth'] = year_pass['Year Of Birth'].astype(str)
for i in year_pass["Year Of Birth"]:
    year_pass.replace(to_replace = i, value = i+str("st"), inplace = True)
```

```python
trace1 = go.Scatter(x=year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"],
                    y=year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"],
                    mode='lines',
                    line=dict(width=2.5), visible = True)

trace2 = go.Scatter(x = year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"],
                    y = year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"],
                    mode='lines',
                    line=dict(width=2.5), visible = True)


frames = [dict(data= [dict(type='scatter',
                           x=year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"][:k+1],
                           y=year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"][:k+1]),
                      dict(type='scatter',
                           x=year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"][:k+1],
                           y=year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"][:k+1])],
               traces= [0, 1],  #
              )for k  in  range(1, len(year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"])-1)] 


layout = go.Layout(width=1000,
                   height=600,
                   showlegend=False,
                   hovermode='closest',
                   updatemenus=[dict(type='buttons', showactive=True,
                                y=1.05,
                                x=1.15,
                                xanchor='right',
                                yanchor='top',
                                pad=dict(t=0, r=10),
                                buttons=[dict(label='Play',
                                              method='animate',
                                              args=[None, 
                                                    dict(frame=dict(duration=10, 
                                                                    redraw=False),
                                                         transition=dict(duration=0),
                                                         fromcurrent=True,
                                                         mode='immediate')])])])


layout.update(xaxis =dict(range=[year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"][0], 
                                 year_pass.sort_values(by=["Year Of Birth"])["Year Of Birth"]
                                 [len(year_pass.sort_values(by=["Year Of Birth"]))-1]], autorange=True),
              yaxis =dict(range=[min(year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"])-0.5, 
                                 year_pass.sort_values(by=["Year Of Birth"])["Total Models Tested"].max()+10000], 
                          autorange=False));
fig = go.Figure(data=[trace1, trace2],frames = frames, layout=layout)
fig.update_layout(title_text='Analysis of Number of models tested with respect to their year of Built',
                 xaxis_title = "Year Of Birth",
                 yaxis_title = "Total Vehicles Tested")
fig.show()
fig.write_html("Test_vs_year.html")
```

# Analysis of Major Fail Domain of Tests carried out

```python
x=[]
for column in (mmd.columns):
    x.append(column)
x=x[8:]
x = [x for i, x in enumerate(x) if i % 2 == 0]
fail = pd.DataFrame(columns=['Fail Domain','Total Vehicles Failed',])
xx=[]
for column in x:
     xx.append(sum(mmd[column]))
```

```python
fail["Total Vehicles Failed"] = xx
fail["Fail Domain"] = x
fail.sort_values(by=["Total Vehicles Failed"], inplace = True)
```

```python
colors = ['black',] * 14
colors[13] = 'crimson'

fig = go.Figure(data=[go.Bar(
    x=fail["Fail Domain"],
    y=fail["Total Vehicles Failed"],
    marker_color=colors # marker color can be a single color value or an iterable
)])
fig.update_layout(title_text='Total Vehicles failed in Different Tests',
                 xaxis_title = "Fail Domain",
                 yaxis_title = "Total Vehicles Failed")
fig.write_html("Total_fail.html")
fig.show()
```
